```python
data = pd.read_csv("results.csv")
data.columns=("X","Y")
print(data)
print(data["Y"].size)

print("Average :",np.mean(data.iloc[:,1]))
print("StDev.  :",np.std(data["Y"]))
```

```python
avg = np.sum(data.iloc[:,1])
# for val in data.iloc[:,1]:
#     avg = avg + val
avg = avg / data["Y"].size
print(avg)

fig1, ax = plt.subplots(figsize=(6,4))

ax.scatter(data["X"],data["Y"])
ax.axhline(y=avg, color='r', linestyle=':')

fig1.show()
```

```python
histo , bins = np.histogram(data["Y"],bins=50)
fig2 , ax = plt.subplots(figsize=(6,4))

ax.bar(bins[:-1],histo,width=0.1,edgecolor='black')
fig2.show()
```
