```python
# python packages
from sys import stdout
import pandas as pd # Dataframes and reading CSV files
import numpy as np # Numerical libraries
import matplotlib.pyplot as plt # Plotting library
from lmfit import Model # Least squares fitting library
```

```python
# Define the figure's parameters
size=30
params = {'figure.figsize': (16,8),
          'legend.fontsize': size*0.7,
          'axes.labelsize': size,
          'axes.titlesize': size,
          'xtick.labelsize': size*0.8,
          'ytick.labelsize': size*0.8,
          'xtick.direction': "out",
          'ytick.direction': "out",
          'axes.linewidth' : 3,
          'axes.titlepad': 25}

from matplotlib.ticker import (MultipleLocator, FormatStrFormatter,
                               AutoMinorLocator)
def prettyTicks(ax):
    
    ax.xaxis.set_ticks_position('both')
    ax.yaxis.set_ticks_position('both')


    ax.xaxis.set_major_locator(MultipleLocator(500))
    ax.yaxis.set_major_locator(MultipleLocator(5))
    
    ax.yaxis.set_minor_locator(AutoMinorLocator())
    ax.xaxis.set_minor_locator(AutoMinorLocator())

    ax.tick_params(which='minor', length=10, width=3, color='black')
    ax.tick_params(which='major', length=20, width=3, color='black')
```

```python
# Import the data
data = pd.read_csv("random2.csv")
data.columns = ("X","Y")
nval = len(data.index)
average = np.mean(data["Y"])

# Set the figure parameters
plt.rcParams.update(params)

# Create the figure with the data and their average as an horizontal line
fig , ax = plt.subplots()
ax.scatter(data["X"],data["Y"],label="Data")
ax.axhline(y=average, color='red', lw=5, label="Average")

# Set the axes labels
ax.set(xlabel="Time (s)")
ax.set(ylabel="Pressure (bar)")

# Set the limits of the axes
plt.xlim([0,nval])
plt.ylim([10,30])

prettyTicks(ax)
    
# Add the legend
ax.legend()

# Make the figure
plt.show()

# Save the figure to file
# plt.savefig("figure.png")
```
